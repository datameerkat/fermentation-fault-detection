## Distribution of parameters
This notebooks analyses the distribution of the parameters in a dataset.

```python
from fermfaultdetect.visualizations import visualize
from fermfaultdetect.utils import get_root, get_simulation_dir
import os
import pandas as pd
from fermfaultdetect.data import statistics, utils
```

### Select dataset

```python
# Change set name here
set_name = "Test_git"

set_dir = os.path.join(get_simulation_dir(), set_name)
batch_list, legend_list = utils.load_batchfolder(set_dir)
```

```python
# Load dataset
batch_list, _ = utils.load_batchfolder(set_dir)

# Calculate statistics
stats_df = statistics.aggregate_stats(batch_list)
```

### Visualization

```python
show_parameters = ["DO_saturation", "weight", "F_set", "air_L_set", "air_L", "kLa", "OUR"]
thesis_colors = visualize.get_thesis_colors()
visualize.set_plot_params(high_res=True)
visualize.plot_batchset_statistics(stats_df, show_parameters, show_std=True, show_minmax=True, separate_plots=True, color=thesis_colors["green"], dpi=300)
```

### Parameter distributions

```python
visualize.plot_parameter_distribution(batch_list, 'E', 230, bins = 20)
visualize.plot_parameter_distribution(batch_list, 'weight', 230, bins = 20)
```
